```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

#### Creamos nuestro producer que va a ser nuestro objecto que va a estar conectado a la EC2 mandando lo que deseemos
###### En la parte donde iria nuestra ip otorgada por AWS esta puesto {ip_public_in_the_aws} por temas de seguridad

```python
producer = KafkaProducer(bootstrap_servers=['{ip_public_in_the_aws}:9092'],
                        value_serializer=lambda x: 
                        dumps(x).encode('utf-8'))
```

```python
producer.send('demo_test', value="{'lobby' : '1'}")
```

```python
df = pd.read_csv("indexProcessed.csv")
```

```python
while True:
    dict_stock = df.sample(1).to_dict(orient="records")[0]
    producer.send('demo_test', value=dict_stock)
    sleep(1)
```
